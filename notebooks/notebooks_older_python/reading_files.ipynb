```python
import testing_functions.reading_files as read
import utilities.benchmark as bench
from utilities.compare_benchmarks import compare_benchmarks, compare_statistics_benchmarks

from config import READING_FILES_ITERATION_COUNT

bench.ITERATIONS = READING_FILES_ITERATION_COUNT
```

```python
SMALL = [
    './test_files/small_datasource.csv',
    './test_files/small_number.csv',
    './test_files/small_string.csv'
]

MEDIUM = [
    './test_files/medium_datasource.csv',
    './test_files/medium_number.csv',
    './test_files/medium_string.csv'
]

BIG = [
    './test_files/big_datasource.csv',
    './test_files/big_number.csv',
    './test_files/big_string.csv'
]
```

```python
read_big_data_benchmark = bench.benchmark(
    read.load_big_data,
    BIG
)

read_medium_data_benchmark = bench.benchmark(
    read.load_medium_data,
    MEDIUM
)

read_small_data_benchmark = bench.benchmark(
    read.load_small_data,
    SMALL
)




read_big_data_polars_benchmark = bench.benchmark(
    read.load_big_data_polars,
    BIG
)

read_medium_data_polars_benchmark = bench.benchmark(
    read.load_medium_data_polars,
    MEDIUM
)

read_small_data_polars_benchmark = bench.benchmark(
    read.load_small_data_polars,
    SMALL
)




read_big_data_chunked_benchmark = bench.benchmark(
    read.load_big_data_chunked,
    BIG
)

read_medium_data_chunked_benchmark = bench.benchmark(
    read.load_medium_data_chunked,
    MEDIUM
)

read_small_data_chunked_benchmark = bench.benchmark(
    read.load_small_data_chunked,
    SMALL
)
```

```python
compare_benchmarks(
    read_big_data_benchmark, 
    read_big_data_polars_benchmark,
    read_big_data_chunked_benchmark
)
compare_statistics_benchmarks(
    read_big_data_benchmark, 
    read_big_data_polars_benchmark,
    read_big_data_chunked_benchmark
)
```
